# **WHILE**

```python
i = 1
while i < 11:
  print(i)
  i += 1
```

```python
i = 1
while i < 11:
  if(i == 5):
    break
  print(i)
  i += 1
```

# **Nested Loop**

```python
for i in range (3):
  for j in range(5):
    print(i,j)
```

```python
#Task

def rank_students(students, exclude_names):
  student_result = []
  rank = 1
  for student in students:
    if student["name"] in exclude_names:
      continue

    if student["score"] == 0:
      break

    if student["score"] >= 50:
      student_result.append({
          "name": student["name"],
          "score": rank
      })
      rank += 1
  return student_result

if __name__ == "__main__":
    students = [
        {"name": "Venki",    "score": 95},
        {"name": "Karthik",  "score": 72},
        {"name": "Harini",   "score": 45},
        {"name": "Abishek",  "score": 88},
        {"name": "Vicky",    "score": 60},
        {"name": "Rinky",    "score":  0},
        {"name": "Surya",    "score": 55},
    ]
    exclude_names = ["Harini"]
    print(rank_students(students, exclude_names))
```
