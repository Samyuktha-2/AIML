# **Types of Arguments**

**Default Parameter**

```python
def my_country(country = "India"):
  print("Hii,I am from ", country)

my_country("Aussie")
my_country()
```

**Keyword Parameter**

```python
def my_animal(animal,animal_name):
  print("I have a animal: ", animal)
  print("My ", animal, "'s name is ", animal_name)


my_animal("Leo","Cat") #wrong as arg order changed
print()
my_animal(animal_name="Leo",animal="Cat")
```

**Positional Argument**

```python
def my_detail(name,age,city):
  print("I am", name,",age is",age,",from",city)


my_detail("Sam",21,"Tup")
```

```python
#Positional argument and keyword argument can be mixed, but positional argument must come before keyword argument

def my_detail(name,age,city):
  print("I am", name,",age is",age,",from",city)

#my_detail(name="Sam","Tup",age=21)  #-->Syntax error
my_detail("Sam",city="Tup",age=21)

#my_detail(21,name="Sam",city="Tup") #-->this also wrong as the P1 take A1, when P1 is assigned with Argument it throws type error
```

**Positional only arguments**

```python
def my_details(name,city,/):
  print("Hii",name,"from",city)

my_details("Sam","Cbe")

#my_details("Sam",city = "Cbe") #-->not possible only position keywords allowed, type error
```

**Keyword only argument**

```python
def my_details(*,name,city):
  print("Hii",name,"from",city)

my_details(name = "Sam",city = "Cbe")

#my_details("Sam","Cbe") #--> only keyword argument is allowed, positional arguments causes error, type error
```

**Combination of Keyword and positional argument**

```python
def my_func(a,b,/,c,*,d,e):
  print(a + b + c + d + e)

# def my_func(*,a,b,c,d,e,/):--> this is incorrect

my_func(1,2,c=3,d=4,e=5)
my_func(1,2,3,d=4,e=5)
```

**Arbitary Positional Argument**

```python
def calculate_bill(name,*price):
  print(price)
  print(type(price))

calculate_bill("Tony",210,120,450)
```

**Arbitrary Keyword Argument**

```python
def calculate_bill(name,*amount):
    bill_amount = 0
    for amt in amount:
        bill_amount += amt
    return bill_amount

print(calculate_bill("Ana",120,85,200))
```
