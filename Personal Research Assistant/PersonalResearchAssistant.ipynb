```python
pip install groq
from groq import Groq
from google.colab import userdata

# Load API key securely from Colab Secrets
api_key = userdata.get("ChatBot")

# Initialize Groq client
client = Groq(api_key=api_key)

# 🔹 Input topic
topic = "Artificial Intelligence in healthcare"

# 🔹 Structured prompt
prompt1 = f"""
Analyze the topic: {topic}

Structure your response as follows:

1. Overview – Explain the concept in simple terms
2. Key Insights – List five major points with brief explanations
3. Did You Know? – Share one interesting or uncommon fact
4. Future Outlook – Describe how this topic might evolve in the future

Keep the tone informative and easy to understand.
"""

# 🔹 Generate response
response1 = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "user", "content": prompt1}
    ]
)

# 🔹 Print output
print("📌 Research Output:\n")
print(response1.choices[0].message.content)

#Story Based Explanation
prompt2 = f"""
Explain '{topic}' by turning it into a short story. Make it engaging and easy to follow.
"""

response2 = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "user", "content": prompt1}
    ]
)

print("\n\n📌 Story Output:\n")
print(response2.choices[0].message.content)
```
