```python
from transformers import pipeline
from groq import Groq
from google.colab import userdata

# Step 1: Load image classification model
classifier = pipeline("image-classification")

# Step 2: Classify image
image_path = "/content/beach.jpg"
result = classifier(image_path)

# Step 3: Format result nicely
object_details = []
for item in result:
    label = item['label']
    score = item['score'] * 100
    object_details.append(f"{label} ({score:.2f}%)")

classified_objects_str = ", ".join(object_details)

print("Detected Objects:")
print(classified_objects_str, "\n")

# Step 4: Load API key securely from Colab Secrets
api_key = userdata.get("ImageClassification")

# Step 5: Initialize Groq client
client = Groq(api_key=api_key)

# -------------------------------
# 🔹 1. Improved Description
# -------------------------------
prompt1 = f"""
Based on the detected objects below, write a creative and engaging description of the image in 2 to 3 lines.

Objects:
{classified_objects_str}

Make the description vivid, natural, and easy to imagine.
"""

response1 = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{"role": "user", "content": prompt1}]
)

print("📌 Description of Image:")
print(response1.choices[0].message.content, "\n")


# -------------------------------
# 🔹 2. Funny Description
# -------------------------------
prompt2 = f"""
Describe this image in a funny way based on these objects:

{classified_objects_str}
"""

response2 = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{"role": "user", "content": prompt2}]
)

print("😂 Funny Description:")
print(response2.choices[0].message.content, "\n")


# -------------------------------
# 🔹 3. Story Description
# -------------------------------
prompt3 = f"""
Describe this image as if you are telling a short story.

Objects:
{classified_objects_str}
"""

response3 = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{"role": "user", "content": prompt3}]
)

print("📖 Story Description:")
print(response3.choices[0].message.content, "\n")
```
