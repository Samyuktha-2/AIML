```python
stream_a = [10, 8, 6, 4, 2]
stream_b = [9, 7, 5, 3, 1]
stream_a.sort()
stream_b.sort()
combined_stream = stream_a + stream_b
combined_stream.sort()
print(combined_stream)
```

```python
inventory = ["iron sword", "shield", "potion", "ration", "shield"]
inventory.append("warhammer")
inventory.insert(0,"elixir")
inventory.extend(["tunic","boots"])
print(inventory)
print("Index of potion: ",inventory.index("potion"))
print("Count of Shield: ",inventory.count("shield"))
inventory.remove("shield") #removes first occurance of shield
print("The last value in inventory: ",inventory.pop())
inventory.sort()
print(inventory)
#inventory.sort(reverse = "true")
inventory.reverse()
print(inventory)
inventory.clear()
print(inventory)
```

```python
#Tuple
tuple1 = ()
print(tuple1)
print(type(tuple1))

tuple1=(1)
print(tuple1)
print(type(tuple1))

tuple1 = (1,)
print(tuple1)
print(type(tuple1))

tuple1=("abc")
print(tuple1)
print(type(tuple1))

tuple1=("abc",)
print(tuple1)
print(type(tuple1))
```
