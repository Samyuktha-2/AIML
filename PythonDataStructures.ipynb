```python
P = {'A','B','C','D'}

J = {'C','D','E','F'}

#SUBSET
print(P.issubset(J))
print(J.issubset(P))

#SUPERSET
print(P.issuperset(J))
print(J.issuperset(P))

#DIFFERENCE
print(P - J)
print(J - P)

#SYMMETRIC DIFFERENCE
print(P ^ J)

#UNION
P |= J
print(P)
print(J)

#INTERSECTION
P = {'A','B','C','D'}
P &= J
print(P)
print(J)


```

```python
#Here is an end-to-end example. You have a list representing the health status of different servers in a data center.
#You want to automate a check-up. If a server is "Offline", you reboot it. If a server status is "Unknown", you skip it.
#If a server reports a "Critical Core Failure", you immediately halt the entire automated process to prevent a fire.

server_statuses = ["Online", "Offline", "Unknown", "Online", "Critical Core Failure", "Offline"]

for status in server_statuses:
  if(status == "Online"):
    print("Server is running smoothly")
  elif (status == "Offline"):
    print("Server is rebooting")
  elif(status == "Unknown"):
    print("Skip the server")
  elif(status == "Critical Core Failure"):
    print("DANGER: Automated server process is halted")
print("\nServer diagnosis completed")
```

```python
def clean_tags(raw_tags):
    cleaned = set(raw_tags)
    return cleaned

def check_required_subset(cleaned, required):
    return required.issubset(cleaned)

def make_frozen_hashable(cleaned):
    try:
        frozen_tags = frozenset(cleaned)
        hash_value = hash(frozen_tags)
        return hash_value
    except TypeError as e:
        print(f"TypeError caught while hashing frozenset: {e}")
        return None

def demo_remove_vs_discard(cleaned, missing_tag):
    try:
        cleaned.remove(missing_tag)
    except KeyError:
        print(f"KeyError caught: '{missing_tag}' is not in the set")

    cleaned.discard(missing_tag)
    print(f"discard('{missing_tag}') completed without error")

if __name__ == "__main__":

  raw_tags = ["python", "data", "python", "ml", "data", "ai", "python", (1, 2), (1, 2), "ml"]

  required_tags_1 = {"python", "data"}
  required_tags_2 = {"python", "java"}

  cleaned = clean_tags(raw_tags)
  print("Unique tag count: ", len(cleaned))
  print("required_tags_1 is a subset: ", check_required_subset(cleaned, required_tags_1))
  print("required_tags_2 is a subset: ", check_required_subset(cleaned, required_tags_2))
  print("frozenset hash: ",make_frozen_hashable(cleaned))
  demo_remove_vs_discard(cleaned, "java")
```
