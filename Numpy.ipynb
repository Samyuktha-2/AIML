**To check time difference**

```python
import time as t
import numpy as np

size = 1000000
python_list = list(range(size))
numpy_array = np.arange(size)

start_time = t.time()
result_list = []
for x in python_list:
  result_list.append(x ** 2)
list_duration = t.time() - start_time
print(f"Python list time: {list_duration:.5f} sec")


start_time = t.time()
array_result = numpy_array ** 2
array_duration = t.time() - start_time
print(f"Python list time: {array_duration:.5f} sec")

speedup = list_duration/array_duration
print(speedup)
```

**Masai Task**

```python
import numpy as np

grades = np.array([
    [80, 90, 85],
    [95, 85, 82],
    [60, 70, 65],
    [88, 92, 79]
])

subject_averages = np.mean(grades, axis = 0)

student_averages = np.mean(grades,axis = 1)

high_performers = grades[student_averages >= 80]

reshaped = grades.reshape(2,6)


if __name__ == "__main__":
    print("Subject averages:", subject_averages)
    print("Student averages:", student_averages)
    print("High performers:\n", high_performers)
    print("Reshaped array:\n", reshaped)
```

# **Filtering**

Filtering in 1D Array

```python
marks = np.array([50,45,84,53,30,75,58])
print(marks)
```

```python
pass_mark = marks > 50
print(pass_mark)

pass_mark = marks[marks > 50]
print(pass_mark)
```

```python
print(marks > 50)
print(marks[marks > 50])
```

```python
#Take even numbers from marks
even_nums = marks[marks % 2 == 0]
print(even_nums)
```

```python
#Print marks above than 50 and below than 70
marks_and = marks[(marks >= 50) & (marks <= 70)]
print(marks_and)
```

```python
#Print marks above than 50 or even mark
marks_or = marks[(marks >= 50) | (marks % 2 == 0)]
print(marks_or)
```

Filtering in Multi-Dimensional Array

Multi dimensional arrays get flatterned and prints the output

```python
arr = np.array([
    [1,2,3],
    [4,5,6]
])
print(arr)
```

```python
even = arr[arr % 2 == 0]
print(even)

even_1 = arr[(arr % 2 == 0) | (arr == 1)]
print(even_1)
```

To avoid 1D output,we can use WHERE
np.where(condition, arrayVariable, value_to_be_printed_if_value_fails)

```python
even = np.where(arr % 2 == 0, arr, 0)
print(even)
```

Can be used in 1D Array also

```python
marks = np.array([50,45,84,53,30,75,58])
even = np.where(marks % 2 == 0, marks, 0)
print(even)
```

**Linear Algebra**

```python
a =np.array([
    [1,2],
    [3,4]
])

b =np.array([
    [2,2],
    [3,4]
])

print(np.matmul(a,b),"\n")  #row 1 col 1, row 1 col 2, row 2 col 1, row 2 col 2

print(a * b, "\n")

print(np.dot(a,b),"\n")

#Transpose
print(b.T,"\n")

#determinant
print(np.linalg.det(b))
```
